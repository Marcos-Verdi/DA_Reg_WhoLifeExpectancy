The missing observations are not exclusive to a certain country or region. We are missing different data accross most of the countries listed.

```python
num_cols = []
cat_cols = []

for col in df.columns:
    if df[col].dtype == 'object':
        cat_cols.append(col)
    else:
        num_cols.append(col)
```

```python
for col in num_cols:
    col_mean = df[col].median()
    df[col].fillna(col_mean, inplace=True)
```

```python
life_expectancy_by_country = df.groupby('Country').mean()[num_cols[1:2]].sort_values(by=num_cols[1:2],ascending=False).head(20)
life_expectancy_by_country
```

As expected, developed countries with high incomes are the ones with higher life expectancies.

## Disease Analysis

```python
def means_by_country(dataframe,col):
    temp = dataframe.groupby("Country").mean()[col].sort_values(ascending=False).head(25)
    print(col,temp)
```

```python
for col in illness_columns:
    means_by_country(df,col)
```

```python
df[df['Country'] == 'Monaco']
```

```python
for i in illness_columns:
    plt.figure(figsize=(20,5))
    sns.scatterplot(x=df['GDP'],y=df[i])
    plt.title(i)
```

As we expected, we see a positive relationship in between GDP per capita and immunization coverage. Likewise we observe a negative relationship between occurrences of HIV and Measles and GDP per capita

```python
for i in illness_columns:
    plt.figure(figsize=(20,5))
    sns.scatterplot(x=df['Income composition of resources'],y=df[i])
    plt.title(i)
```

In the case of income composition resources, we see that it has a positive relationship with immunization coverage and a negative one with occurrences of specific illnesses. However, in the case of HIV we can observe that the peak of occurrences occurs in countries where the income composition is about 0.5 but as expected remains ver low as that percentage gets higher

```python
for i in illness_columns:
    plt.figure(figsize=(20,5))
    sns.scatterplot(x=df['Total expenditure'],y=df[i])
    plt.title(i)
```

Also, as expected, we see normal correlations between government health expenditure and immunization coverage and illnesses occurrences. However, the data also shows that many governments could improve their efficiency in terms of expenditure given that we see governments spending the same proportion of their GDP but obtaining different results.

```python
for i in illness_columns:
    plt.figure(figsize=(20,5))
    sns.scatterplot(x=df['Schooling'],y=df[i])
    plt.title(i)
```
